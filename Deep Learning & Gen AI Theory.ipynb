# Diff Between ANN, CNN & RNN

*  `Artificial Neural Network (ANN):`

    - Also known as a Feedforward Neural Network (FNN), it's the most basic form of neural network.
    - It consists of an input layer, one or more hidden layers, and an output layer. Neurons in adjacent layers are fully connected.
    - Suitable for a wide range of tasks, including classification, regression, and simple pattern recognition.
    - Lacks the ability to handle sequential or spatial data efficiently.

* `Convolutional Neural Network (CNN):`

    - Primarily designed for image and video recognition tasks.
    - Employs convolutional layers, which use filters (also called kernels) to automatically learn and identify features from input data.
    - Contains convolutional layers followed by pooling layers to reduce spatial dimensions and extract hierarchical features.
    - Well-suited for tasks involving grid-like data, where spatial relationships are important. It reduces the number of parameters and learns hierarchical representations efficiently.
    - Used in image classification, object detection, image segmentation, and more.

* `Recurrent Neural Network (RNN):`

    - Designed to handle sequential data, such as time series, text, and speech.
    - Contains connections that loop back on themselves, allowing the network to maintain internal memory.
    - Each time step in the sequence processes an input and the hidden state from the previous step.
    - Common issue: vanishing gradient problem, which can make it difficult to capture long-range dependencies in sequences.
    - Used in tasks like language modeling, machine translation, speech recognition, and more.
    - More advanced RNN variations, like Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU), have been developed to mitigate the vanishing gradient problem and improve learning of long-term dependencies.

## RELU - Rectified Linear Unit

`The ReLU (Rectified Linear Unit) activation function is a popular choice for dense layers in neural networks for a variety of tasks, including text summarization`.     
- Once we have built the dense layer, that our inputs are sent to the hidden layers there is a way to imprpove
process and result of these dense layers using something called as RELU.
__the ReLU function gives an output of 0 if the input is negative or zero,
and if input is positive, then the output will be equal to the input__    
These activation functions serve 2 purpose:    


1.   <font color = 'Purple'>Help a model account for non-linear effects</font>-- non linear is where data cannot be sepearated linearly, example if bank churn data has multiple variables it cannot be seperated into 2 class linearly.
    - Nonlinear transformations are applied within activation functions to capture complex relationships in data that cannot be represented by simple linear transformations.
2.   <font color= 'purple'>Help a model account for interaction effects i.e. if A affects the probability of prediction given value of B.    
1. ReLU is a computationally efficient activation function, which makes it well-suited for training large neural networks.

__<font color='DC143C'> Some of the other important activation functions include Sigmoid, tanh, ELU</font>__    

*   **Flattening:** The process of converting a 2d image into 1d vecto
*   **Softmax:** A function that provides probabilities for each possible output class
    
`ReLU helps to overcome the vanishing gradient problem encountered with the sigmoid (and also tanh) function in a few ways:`

__Non-saturating: Unlike the sigmoid function, ReLU does not saturate in the positive region. The gradient for ReLU in the positive region is always 1, no matter how large the input. This means during backpropagation, when gradients are being passed back through the network, the gradients do not shrink to very small values as with sigmoid functions. This allows the weights of the network to be updated more effectively, even for neurons that are far from the output layer.__

## How does RELU gives 0 or 1 as output

- Lets assume we are working on the dataset as # Generate some sample input data input_data = [-2, -1, 0, 1, 2]
- For input values -2 and -1 (negative), the ReLU activation function will produce an output of 0, as stated in the ReLU behavior.
- For input value 0, the ReLU activation function will also produce an output of 0.
- For input values 1 and 2 (positive), the ReLU activation function will pass the input value as the output value, as stated in the ReLU behavior.

## How to Preserving the original scale
Suppose we are working on a customer sentiment problem where the scale goes from -5 to 5 where -5 is highly dissatisfied, 0 is neutral and 5 is extermely satisfied.  if RELU changes -5 to a 0 in the processing how would the data scientist know the impact and severity

* The behavior of the ReLU activation function, where negative values are turned into 0, may not be suitable for certain types of data that have meaningful values outside the non-negative range
* To address this issue, you could consider the following approaches:

    - Activation Function Selection: Instead of using ReLU directly on the output layer, you might opt for an activation function that can map the network's output to the desired sentiment scale. For example, you could use a custom activation function or a modified version of ReLU that is adapted to the specific sentiment scale.

    - Normalization and Rescaling: If you still choose to use ReLU or similar activation functions, you could preprocess your target labels (sentiment values) to a non-negative range (e.g., 0 to 10) before training the model. This way, the ReLU activation function wouldn't cause any values to be lost, and you could post-process the model's predictions to map them back to the original -5 to 5 scale.

    - Loss Function Design: You might also consider designing a custom loss function that takes into account the scale and meaningfulness of your sentiment values. This could help guide the training process to pay more attention to the severity of sentiment changes.

    - Output Post-Processing: Regardless of the chosen activation function, it's a common practice to post-process the model's predictions to ensure they are meaningful in the context of your application. For example, you could apply rounding, scaling, or other transformations to align the predictions with the -5 to 5 sentiment scale.

    - Benefits of ROPE:

      - Relative Position Encoding: The rotation itself inherently encodes relative positions between words. Words closer in the sequence will have more similar rotations compared to words further apart.
      - Efficiency: ROPE computations involve matrix multiplications, which can be more efficient than learning and adding separate embedding vectors

1. __Data and Semantic Information__ Zero-shot learning begins with a dataset that includes labeled examples of some classes (known classes) and a description or attributes associated with each class.
These descriptions or attributes can be in the form of textual descriptions, vectors, or any other structured information that captures the characteristics of the classes
1. __Feature Extraction:__ The labeled examples of known classes are used to train a feature extraction model. This model learns to extract meaningful features from the data, which can be used to represent images, text, or other data types
1. __Semantic Embeddings:__ The textual descriptions or attributes associated with each class are used to create semantic embeddings. These embeddings represent the semantic relationships between classes. Common methods include word embeddings (e.g., Word2Vec, GloVe) for text-based descriptions
1. __Model Training:__ A model is trained to map the extracted features and semantic embeddings into a shared space. This shared space allows the model to associate features with their corresponding class semantics.
Various machine learning algorithms can be used, such as neural networks, support vector machines (SVMs), or other classifiers
1.__Zero-Shot Prediction:__ To make predictions for unseen classes, you provide the model with the features of an example (e.g., an image or text) and the semantic embedding or description of the target class.
The model then computes a similarity or compatibility score between the features and the class semantics for all possible classes.

# Few Shot Learning
You can use the existing knowledge of the LLM and fine-tune it with the few available examples of the new category. This allows the LLM to adapt and learn the specific characteristics of the new category, even though it has not seen many examples.

Here's how the process might work:

1. **Provide the LLM with the few labeled examples of the new category.** For instance, if the new category is "scientific articles about climate change," you would provide the LLM with a handful of scientific articles labeled as belonging to this category.
2. **Fine-tune the LLM on these few examples.** This involves adjusting the parameters of the LLM to improve its performance specifically on the new category.
3. **Test the LLM on new, unseen texts.** After fine-tuning, the LLM can be used to classify new texts into the new category, even though it was only trained on a few examples.

# Vector Library
1. FAISS -
- FAISS is a library for efficient similarity search and clustering of dense vectors
- It is primarily designed for high-dimensional data and is optimized for GPUs.
- FAISS allows indexing of large vectors, and it can handle billions of records efficiently.
- It works by constructing an index that stores vectors in a compressed format to save memory and speed up queries.
- The library provides algorithms for performing similarity search and clustering on the stored vectors based on several algorithms, including hierarchical k-means and product quantization.
- It uses a distance metric, such as Euclidean distance or cosine distance, to measure the similarity between vectors.
- FAISS can handle multiple indexes simultaneously, allowing users to perform several searches with different configurations.
- The library includes both CPU and GPU versions that support different hardware configurations.
- Users can use FAISS through a Python interface or C++ API.
- FAISS is widely used for image and natural language processing applications, including face recognition, recommendation systems, and document similarity searches.

## Distance Metrics [Algorithms]:
- `Manhattan and Euclidean distance`, measure the distance between the endpoints of the vectors.
  - `Manhattan distance`, also known as the L1 distance. Manhattan distance calculates the distance between two points in a grid-like pattern, like the streets of Manhattan in New York City.
  - It is defined as the sum of the absolute differences between the corresponding coordinates of the two points, where one can only move along one axis at a time.
  - `Euclidean distance`, also known as the L2 or square distance.
  - Euclidean distance calculates the shortest distance between two points in a straight line.
 - It is defined as the square root of the sum of the squared differences between the corresponding coordinates of the two points.
- `Cosine distance`, measures the angle between two vectors and defines the vector similarity in terms of directions.
  - Cosine distance, which measures the similarity of two vectors in terms of their direction.
  - Cosine similarity is calculated by taking the cosine of the angle between the two vectors, where zero indicates that the vectors are completely aligned and one indicates that they are completely orthogonal
- `Dot product distance`, considers the similarity in terms of both the direction and the magnitude of two vectors.
  - Also known as the inner product distance
  - It is based on the projection of one vector onto another
  - It considers the similarity between two vectors in terms of both their direction and magnitude

## RLFS :
- RLHS, or Reinforcement Learning with Human Scoring, is a training method that combines human feedback with reinforcement learning to improve the performance of large language models (LLMs). It addresses the limitations of purely supervised learning, where models are trained on static datasets and may not generalize well to real-world scenarios.
- The LLM is first pre-trained on a massive dataset of text, similar to traditional supervised learning.
- Once the LLM is pre-trained, it is used to generate outputs for specific tasks. These outputs are then evaluated by humans, who provide feedback on their quality and relevance. This feedback can be in the form of ratings, rankings, or even free-form text
- The human feedback is used as a reward signal in a reinforcement learning framework. The LLM interacts with the environment (the task it's trying to perform), receives feedback from humans, and learns to adjust its parameters to maximize the rewards (positive feedback). This iterative process allows the model to learn from its mistakes and improve its performance over time.

## Distilling 

- Extracting Knowledge into a Smaller Model
-Once you've customized an LLM using RLHS or other methods, you might want to distill it into a smaller, more efficient model. __This is particularly useful for deployment on devices with limited resources or for reducing inference time.__
- Choose a smaller model architecture, such as a Transformer with fewer layers or a smaller embedding size. This model will be the "student" that learns from the customized LLM (the "teacher").
- There are two main approaches to distilling knowledge from the teacher to the student:
  - * **Soft Distillation:** The student model learns from the teacher's predicted probability distribution over the vocabulary, not just the final predicted word. This allows the student to capture some of the teacher's uncertainty and learn from its "soft" decisions.
    * **Hard Distillation:** The student model simply learns from the teacher's final predictions (hard labels). This is simpler but might be less effective than soft distillation.


## Adapter Tuning [LORA]

- Adapter tuning is another technique for customizing LLMs that offers a balance between efficiency and performance. Unlike full fine-tuning, which updates all the parameters of the LLM, adapter tuning only adds small, task-specific modules ("adapters") to the existing architecture.
- Adapters are small neural network modules inserted into specific layers of the LLM architecture. These adapters can be placed in different locations, such as within attention layers or feed-forward networks.
- During training, only the parameters of the adapters are updated, while the original LLM parameters remain frozen. This significantly reduces the number of parameters that need to be trained, making adapter tuning more efficient than full fine-tuning.

Benefits of Adapter Tuning:**

* **Parameter efficiency:** Adapter tuning requires significantly fewer parameters to be trained compared to full fine-tuning, making it more efficient and faster.
* **Task-specific customization:** Adapters allow for task-specific customization without affecting the LLM's performance on other tasks.
* **Modular and flexible:** Adapters can be easily added or removed, providing flexibility in customizing the LLM for different needs.

**Challenges of Adapter Tuning:**

* **Performance:** While adapter tuning can be effective, it might not always match the performance of full fine-tuning on all tasks.
* **Placement and design:** The optimal placement and design of adapters can vary depending on the task and LLM architecture.


### Internals LORA:
- LORA (Low-Rank Adaptation) is a technique for efficiently fine-tuning large language models (LLMs) by `freezing the original model weights` and `injecting trainable rank-decomposition matrices into specific layers`. This approach significantly reduces the number of trainable parameters while achieving comparable performance to full fine-tuning.
- Instead of fine-tuning the entire LLM, LORA **freezes the original model weights** and introduces **trainable low-rank matrices (A and B)** into specific layers, such as attention and feed-forward layers. These matrices are typically much smaller than the original weight matrices, significantly reducing the number of trainable parameters.

- Formula:
   * <font color='red'>__max_ _{(x, y)  Z} _{t=1}^{|y|} log(P_(y_t | x_t, y < t))__ </font>
   *  **:** This represents the **parameters** of the model, which in this case is a **policy**. A policy defines the probability of taking each possible action in a given state.
* **(x, y):** This represents a **state-action pair**.
    * **x:** The state of the environment at a given time step.
    * **y:** The action taken in that state.
* **Z:** This is the set of all possible **state-action pairs**.
* **t:** This represents a specific **time step**.
* **|y|:** This denotes the **total number of time steps** in the trajectory (sequence of state-action pairs).
* **P_(y_t | x_t, y < t):** This is the **conditional probability** of taking action **y_t** at time step **t**, given the current state **x_t** and all the previous actions **y < t**. This probability is determined by the policy ****.
* **log(...):** The **logarithm** function is applied to the probability to transform it into a more convenient form for optimization.
* **:** The **summations** aggregate the log probabilities over all state-action pairs in the trajectory and over all time steps.
* **max_:** This indicates that we are searching for the **policy parameters  that maximize** the sum of log probabilities.

**In statistical terms, this formula represents the **maximum likelihood estimation (MLE)** of the policy parameters.** We are trying to find the policy that maximizes the likelihood of observing the given trajectory of state-action pairs. By maximizing this likelihood, we are essentially finding the policy that is most likely to generate successful outcomes (high rewards) in the future.

**This formula is commonly used in **policy gradient methods** in reinforcement learning.** These methods iteratively update the policy parameters in the direction that increases the likelihood of observing desirable trajectories.

# Distilling Vs Adapter Tuning [When To Use]
- Distillation:** If your primary goal is to create a smaller, faster, and more efficient model while preserving most of the original LLM's performance, distillation can be a good choice. * **Adapter tuning:** If you want to customize the LLM for specific tasks without significantly increasing the model size or sacrificing performance on other tasks, adapter tuning is a suitable option.
